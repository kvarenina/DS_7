### Завдання 6 

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
import statistics

#генеруэмо тисячу чисел в діапазоні від 0 до 10
random_list = [random.randint(0,100) for i in range(1000)]

def minmax(random_list):

    print(random_list,'\n')
    minim = min(random_list)
    maxim = max(random_list)
    mean = statistics.mean(random_list)
    med = statistics.median(random_list)
    st_dev = statistics.stdev(random_list)
    print('Мінімальне значення: ', minim)
    print('Максимальне значення: ', maxim)
    print('Середнє значення: ', mean) 
    print('Медіана: ', med)
    print('Середнє квадратичне відхилення: ', st_dev, '\n')    
    
minmax(random_list)


def num_count(random_list):
    counts = dict()
    for num in random_list:
        if num in counts:
            counts[num] += 1
        else:
            counts[num] = 1
    return counts
print('Частота цифр в послідовності: \n',num_count(random_list))

  
                 
```

```python
#знаходим середнє значення
n = len(random_list)
get_sum = sum(random_list)
mean = get_sum/n
print("Mean is: ", mean)

#знаходим медіану
random_list.sort()
if n % 2 == 0:
    median1 = random_list[n//2]
    median2 = random_list[n//2 - 1]
    median = (median1 + median2)/2
else:
    median = random_list[n//2]
print("Median is: " + str(median))

#знаходим середнє квадр. відхил.
variance = sum([((x - mean) ** 2) for x in random_list]) / len(random_list)
stdev = variance ** 0.5
print("Standart deviation is: ", stdev)
        
```
