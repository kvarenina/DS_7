```python
DF = {}
#перетворюємо стрічки у списки, розділяючим знаком 
for line in open('ds_7.txt', encoding='UTF-8'):
    tmp = line.split('|')
#задаємо стовпці як індекси списку    
    key = tmp[0]
    name = tmp[1]
    on_stock = tmp[2]
    price = int(tmp[3].split()[0]) 
#записуємо все у словник    
    DF[key] = [name, on_stock, price]
        
f.close()

DF
```

```python
#знайдемо кількість елементів
len(DF)
```

```python
#кількість товарів у наявності
quantity = 0
for key in DF:
    if DF[key][1] == 'В наличии':
        quantity +=1
quantity
```

```python
#другий варіант
len(list(filter(lambda key: DF[key][1] == 'В наличии', DF)))
```

```python
#немає в наявності
for key in DF:
    if DF[key][1] != 'В наличии':
        print(key, DF[key])     
```

```python
#знайти товар з найбільшою ціною
price = -1
name = None

for key in DF:
    if DF[key][2] > price:
        price = DF[key][2]
        name = DF[key][0]
print(name, price)
```
